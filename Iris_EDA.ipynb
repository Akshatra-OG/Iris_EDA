```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
data = load_iris()

# Create a DataFrame for easier analysis
iris_df = pd.DataFrame(data.data, columns=data.feature_names)
iris_df['target'] = data.target
```

```python
iris_df['target'].unique()
```

```python

# Visualize the distribution of features
plt.figure(figsize=(12, 6))
for i, feature in enumerate(data.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.histplot(iris_df[feature], bins=20, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Pairplot to visualize relationships between features
sns.pairplot(iris_df, hue='target', palette='Set1')
plt.suptitle('Pairplot of Iris Dataset Features', y=1.02)
plt.show()

```

```python
# Box plots to visualize feature distributions by target class
plt.figure(figsize=(12, 6))
for i, feature in enumerate(data.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='target', y=feature, data=iris_df, palette='Set1')
    plt.title(f'{feature} by Target Class')
    plt.xlabel('Target')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()

```

```python
# Correlation matrix
correlation_matrix = iris_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Matrix')
plt.show()
```
